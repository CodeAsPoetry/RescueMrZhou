```python
import os
import re
import json
import uuid
from tqdm.notebook import tqdm
```

```python
corpus_file_path = '../Corpus/'
lyric_file_path_list = os.listdir(corpus_file_path)
lyric_file_path_list
```

```python
music_lyric = {}
author_reg = re.compile(u'词\：方文山')

lyric_fangwenshan_file_path = '../data/fangwenshan_lyrics.json'

samples_list = []
for lyric_file_path in lyric_file_path_list:
    with open(corpus_file_path + lyric_file_path, 'r', encoding='utf-8') as lyric_file:
        lyric_lines = lyric_file.readlines()
        if len(lyric_lines) == 0:
            continue
        lyric_list = []
        for lyric_line in lyric_lines:
            lyric_line = lyric_line.strip()
            if lyric_line != '':
                lyric_list.append(lyric_line)
    
        lyric = '\n'.join(lyric_list)
        result = re.findall(author_reg, lyric)
        if len(result)>0:
            temp_dict = {
                'id' : str(uuid.uuid1()),
                'name': lyric_file_path[:-4],
                'lyric': lyric
            }
            samples_list.append(temp_dict)

            
            
print(len(samples_list))
```

```python
def compute_iou_value(str1, str2):
    return len(set(str1) & set(str2)) / len(set(str1) | set(str2))

has_repeat_pair = []
gank_list = []

for i in range(len(samples_list)):
    for j in range(i+1, len(samples_list)):
        if compute_iou_value(samples_list[i]['lyric'], samples_list[j]['lyric']) > 0.9:
            has_repeat_pair.append([i, j, compute_iou_value(samples_list[i]['lyric'], samples_list[j]['lyric'])])
            gank_list.append(j)
    
    
        
gank_list = list(set(gank_list))
len(gank_list)
```

```python
fang_lyric_file = open(lyric_fangwenshan_file_path, 'w', encoding='utf-8')

for i in range(len(samples_list)):
    if i not in gank_list:
        fang_lyric_file.write(json.dumps(samples_list[i], ensure_ascii=False))
        fang_lyric_file.write('\n')
fang_lyric_file.close()
```
